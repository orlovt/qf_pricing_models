```python
from  analytical_differences import difference as dif
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline

import numpy as np 
```

```python
plt.figure(figsize=(8, 6))
deltasxs = [0.5, 0.1]
nums = np.arange(1,5,0.01)
true = dif.f(nums)

plt.plot(nums, true, label= 'True',linewidth=3)
for delt in deltasxs:
    plt.plot(nums, dif.forward_difference(dif.f, nums, delt), label=f'Forward$\Delta x$ = {delt}',linewidth=2,linestyle='--')
    plt.plot(nums, dif.backward_difference(dif.f, nums, delt), label=f'Backward$\Delta x$ = {delt}',linewidth=2,linestyle=':')
    plt.plot(nums, dif.central_difference(dif.f, nums, delt), label=f' Central $\Delta x$ = {delt}',linewidth=2,linestyle='-.')
    

plt.legend()
plt.ylabel("Derivative")
plt.xlabel('$x$')
plt.title('Forward, Backward and Central Finite Difference')
```
